```python
%pylab inline
from scipy.interpolate import interp1d
from scipy.fftpack import fft, fftfreq
```

```python
#obtenemos los datos
data=loadtxt("RR_Lyrae_template.dat")
size_data=shape(data)
print size_data
chi= empty((11))#chi cuadrado, está debajo de cada gráfica
#número de puntos para hacer lagrange
n_points1=size_data[0]
#x_data corresponde a el tiempo y y_data corresponde al número de manchas solares
x_data=zeros(n_points1)
y_data=zeros(n_points1)

x_data=data[:,0]
y_data=data[:,1]
#Tomamos los datos mayores a cero
index=where(y_data>=0.0)
x_data=(x_data[index])
y_data=(y_data[index])
n_points=len(index)



plot(x_data,y_data)
scatter(x_data,y_data,s=5)

```

```python
#interpolación de lagrange

lagrange = interp1d(x_data, y_data, kind = "linear")

tamano = size(x_data)

tiempo = linspace(x_data[0],x_data[tamano-1],2000)

plot(tiempo,lagrange(tiempo))

```

```python
#Hagamos transformada
transformada = (fft(lagrange(tiempo)) )
freq= (fftfreq(2000, (x_data[tamano-1]-x_data[0])/2000)) # Recuperamos las frecuencias
plt.axis((-12.5,12.5,0,5000))
plt.plot(freq,abs(transformada))

```

```python
orden=sort(abs(transformada))
```

```python
#Si tomamos por ejemplo los 11 mayores valores

mayores=orden[-11::]
size(mayores)
print(mayores)
rango=size(transformada)
#hacemos cero la función en los puntos donde no hay maximos
for i in range(rango):
    valoradefinir=abs(transformada[i])
    if(valoradefinir<min(mayores)):
        transformada[i]=0.0
plt.axis((-12.5,12.5,0,1000))
plot(freq,abs(transformada))
```

```python
cleanf = ifft(transformada)
plot(data[:,0],cleanf,c='b')
plot(tiempo,lagrange(tiempo))
chi[10]=sum((y_data-cleanf)**2)

print abs(sum(y_data-cleanf)**2)
```
