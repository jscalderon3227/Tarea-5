```python
%pylab inline
from scipy.interpolate import interp1d
from scipy.fftpack import fft, fftfreq
```

```python
#obtenemos los datos
data=loadtxt("RR_Lyrae_template.dat")
size_data=shape(data)
print size_data
#número de puntos para hacer lagrange
n_points1=size_data[0]
#x_data corresponde a el tiempo y y_data corresponde al número de manchas solares
x_data=zeros(n_points1)
y_data=zeros(n_points1)

x_data=data[:,0]
y_data=data[:,1]
#Tomamos los datos mayores a cero
index=where(y_data>=0.0)
x_data=(x_data[index])
y_data=(y_data[index])
n_points=len(index)



plot(x_data,y_data)
scatter(x_data,y_data,s=5)

```

```python
#interpolación de lagrange

lagrange = interp1d(x_data, y_data, kind = "linear")

tamano = size(x_data)

tiempo = linspace(x_data[0],x_data[tamano-1],1000)

plot(tiempo,lagrange(tiempo))
scatter(x_data,y_data,s=1)
```

```python
#Hagamos transformada
transformada = fft(lagrange(tiempo) )/ 1000 
freq= real(fftfreq(1000, (x_data[0]-x_data[tamano-1])/1000)) # Recuperamos las frecuencias
plt.plot(freq,real(transformada))

```

```python
orden=sort(abs(transformada))
```

```python
#Si tomamos por ejemplo los 11 mayores valores

mayores=orden[-11::]
size(mayores)
mayores
rango=size(transformada)
#hacemos cero la función en los puntos donde no hay maximos
for i in range(rango):
    valoradefinir=transformada[i]
    if(valoradefinir<min(mayores)):
        transformada[i]=0.0
plot(freq,transformada)
```

```python
#interpolación de lagrange

lagrange2 = interp1d(freq,transformada, kind = "linear")

tamano2 = size(freq)

tiempo2 = linspace(freq[0],freq[tamano2-1],1000)

plot(tiempo2,lagrange2(tiempo2))
scatter(freq,real(transformada),s=3)
```
